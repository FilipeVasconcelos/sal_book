```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def y(Cn,t,w1=0.25):
    x = np.ones(len(t))*np.real(Cn[0])/2
    for k in range(1,len(Cn)):
        x+=abs(Cn[k])*np.cos(k*w1*t+np.angle(Cn[k]))
    return x

def cn_carre(N,a=1,moyenne=0):
    n = np.arange(1,N+1)
    cn = -1j*2*a/(n*np.pi)*(1-(-1)**n)
    return np.concatenate(([2*moyenne],cn)) # moyenne nulle

def cn_triangle(N,a=1,moyenne=0):
    n = np.arange(1,N+1)
    cn = -4*a/(n*np.pi)**2*(1-(-1)**n)
    return np.concatenate(([2*moyenne],cn))

t=np.linspace(0,24*np.pi,2048)
plt.plot(t,y(cn_triangle(128,a=2,moyenne=2),t))
```
